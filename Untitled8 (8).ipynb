```python
# Cell 3: Scrape LangChain 1.0 Documentation

import requests
from bs4 import BeautifulSoup
import time

def scrape_langchain_doc(url):
    """Scrape a single LangChain doc page"""
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            for tag in soup.find_all(['nav', 'footer', 'script', 'style']):
                tag.decompose()
            text = soup.get_text(separator='\n', strip=True)
            lines = [line.strip() for line in text.split('\n') if line.strip()]
            return '\n'.join(lines)
        return None
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None

urls = [
    "https://python.langchain.com/docs/introduction/",
    "https://python.langchain.com/docs/concepts/",
    "https://python.langchain.com/docs/how_to/",
    "https://python.langchain.com/docs/tutorials/",
    "https://python.langchain.com/docs/concepts/agents/",
    "https://python.langchain.com/docs/concepts/rag/",
    "https://python.langchain.com/docs/concepts/chat_models/",
    "https://python.langchain.com/docs/concepts/embeddings/",
    "https://python.langchain.com/docs/concepts/vectorstores/",
    "https://python.langchain.com/docs/concepts/retrievers/",
    "https://changelog.langchain.com/announcements/langchain-1-0-now-generally-available",
    "https://changelog.langchain.com/announcements/langgraph-1-0-is-now-generally-available",
]

documents = []
for url in urls:
    print(f"Scraping: {url}")
    text = scrape_langchain_doc(url)
    if text and len(text) > 200:
        documents.append({
            'url': url,
            'content': text,
            'source': url.split('/')[-2] or url.split('/')[-1]
        })
        print(f"  Done - {len(text)} characters")
    else:
        print(f"  Failed or too short")
    time.sleep(1)

print(f"\nTotal documents scraped: {len(documents)}")
```

```python
# Cell 7: Enhanced RAG Pipeline Functions

from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from rank_bm25 import BM25Okapi
import time

# Initialize LLM
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GEMINI_API_KEY)

# Hybrid Search (BM25 + Vector)
def hybrid_search(query, vectorstore, chunks, k=6):
    # Vector search
    vector_results = vectorstore.similarity_search(query, k=k)

    # BM25 keyword search
    tokenized_corpus = [chunk.page_content.lower().split() for chunk in chunks]
    bm25 = BM25Okapi(tokenized_corpus)
    tokenized_query = query.lower().split()
    bm25_scores = bm25.get_scores(tokenized_query)

    # Get top BM25 results
    top_bm25_indices = sorted(range(len(bm25_scores)),
                               key=lambda i: bm25_scores[i],
                               reverse=True)[:k]
    bm25_results = [chunks[i] for i in top_bm25_indices]

    # Combine and deduplicate
    seen_contents = set()
    combined_results = []
    for doc in vector_results + bm25_results:
        content_key = doc.page_content[:100]
        if content_key not in seen_contents:
            seen_contents.add(content_key)
            combined_results.append(doc)

    return combined_results[:k]

# Enhanced RAG Answer (no query rewriting - saves API calls)
def enhanced_rag_answer(question, vectorstore, chunks):
    # Step 1: Hybrid search
    retrieved_docs = hybrid_search(question, vectorstore, chunks, k=6)

    # Step 2: Cross-encoder reranking
    reranked_docs = rerank_results(question, retrieved_docs, top_k=4)

    # Step 3: Generate answer with context
    context = "\n\n".join([doc.page_content for doc in reranked_docs])

    rag_prompt = ChatPromptTemplate.from_template(
        """You are an expert LangChain documentation assistant.
        Answer the question using ONLY the provided context.
        Be specific and cite exact details from the context.
        If the context does not contain enough information, say so clearly.

        Context:
        {context}

        Question: {question}

        Detailed Answer:"""
    )

    chain = rag_prompt | llm
    response = chain.invoke({
        "context": context,
        "question": question
    })

    return {
        "original_query": question,
        "rewritten_query": question,
        "num_docs_retrieved": len(retrieved_docs),
        "answer": response.content
    }

print("RAG pipeline functions loaded successfully")
print("Advanced techniques: Hybrid Search (BM25 + Vector) + Cross-Encoder Reranking")
```

```python
# Cell 7b: Cross-Encoder Reranker

from sentence_transformers import CrossEncoder

# Load cross-encoder model
reranker_model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')

def rerank_results(query, documents, top_k=4):
    """Rerank documents using cross-encoder"""
    if not documents:
        return documents

    pairs = [[query, doc.page_content] for doc in documents]
    scores = reranker_model.predict(pairs)
    scored = sorted(zip(scores, documents), key=lambda x: x[0], reverse=True)

    return [doc for score, doc in scored][:top_k]

print("Cross-encoder reranker loaded successfully")
print("Model: cross-encoder/ms-marco-MiniLM-L-6-v2")
```

```python
# Cell 9: Comprehensive Evaluation Metrics
# Precision@K, Recall@K, MRR, Grounding, Correctness

import json

# ============================================================
# STEP 1: Define Ground Truth
# Which chunks are RELEVANT for each question
# (Based on keywords that should appear in relevant chunks)
# ============================================================

ground_truth = {
    "What is the new agent creation method in LangChain 1.0?": {
        "relevant_keywords": ["create_agent", "agent", "middleware", "langchain 1.0"],
        "correct_answer_keywords": ["create_agent"],
        "correct_answer": "create_agent"
    },
    "What Python version does LangChain 1.0 require?": {
        "relevant_keywords": ["python", "3.10", "version", "required"],
        "correct_answer_keywords": ["3.10"],
        "correct_answer": "Python 3.10+"
    },
    "What is DeltaChannel in LangGraph?": {
        "relevant_keywords": ["deltachannel", "delta", "checkpoint", "incremental"],
        "correct_answer_keywords": ["delta", "checkpoint", "incremental"],
        "correct_answer": "stores incremental delta, cuts checkpoint overhead"
    },
    "What changed from LangChain 0.3 to 1.0?": {
        "relevant_keywords": ["migration", "breaking", "create_agent", "3.10", "langchain-classic"],
        "correct_answer_keywords": ["create_agent", "3.10", "langchain-classic"],
        "correct_answer": "Python 3.10+, create_agent, langchain-classic package"
    },
    "Does LangChain 1.0 support Python 3.9?": {
        "relevant_keywords": ["python", "3.9", "dropped", "3.10", "required"],
        "correct_answer_keywords": ["no", "dropped", "3.10"],
        "correct_answer": "No - Python 3.9 support was dropped"
    }
}

# ============================================================
# STEP 2: Calculate Retrieval Metrics
# ============================================================

def is_relevant(chunk_content, relevant_keywords):
    """Check if chunk contains relevant keywords"""
    content_lower = chunk_content.lower()
    return any(kw.lower() in content_lower for kw in relevant_keywords)

def precision_at_k(retrieved_chunks, relevant_keywords, k):
    """Precision@K = relevant retrieved / k"""
    top_k = retrieved_chunks[:k]
    relevant_count = sum(1 for c in top_k if is_relevant(c.page_content, relevant_keywords))
    return relevant_count / k if k > 0 else 0

def recall_at_k(retrieved_chunks, relevant_keywords, k, total_relevant):
    """Recall@K = relevant retrieved / total relevant"""
    top_k = retrieved_chunks[:k]
    relevant_count = sum(1 for c in top_k if is_relevant(c.page_content, relevant_keywords))
    return relevant_count / total_relevant if total_relevant > 0 else 0

def mrr_score(retrieved_chunks, relevant_keywords):
    """MRR = 1/rank of first relevant chunk"""
    for i, chunk in enumerate(retrieved_chunks, 1):
        if is_relevant(chunk.page_content, relevant_keywords):
            return 1.0 / i
    return 0.0

def grounding_score(answer, retrieved_chunks):
    """Check if answer content appears in retrieved chunks"""
    answer_words = set(answer.lower().split())
    chunk_words = set()
    for chunk in retrieved_chunks:
        chunk_words.update(chunk.page_content.lower().split())
    overlap = len(answer_words.intersection(chunk_words))
    return min(overlap / max(len(answer_words), 1), 1.0)

def correctness_score(answer, correct_keywords):
    """Check if answer contains correct answer keywords"""
    answer_lower = answer.lower()
    found = sum(1 for kw in correct_keywords if kw.lower() in answer_lower)
    return found / len(correct_keywords) if correct_keywords else 0

# ============================================================
# STEP 3: Run Evaluation
# ============================================================

print("EVALUATION RESULTS")
print("=" * 70)
print("Comparing Baseline LLM vs Enhanced RAG System")
print("=" * 70)

K = 6  # Number of chunks retrieved
evaluation_results = []

for question in test_questions:
    gt = ground_truth[question]
    baseline_ans = baseline_results.get(question, "")
    rag_data = rag_results.get(question, {})
    rag_ans = rag_data.get("answer", "")

    # Get retrieved chunks for this question
    retrieved = hybrid_search(question, vectorstore, chunks, k=K)
    reranked = rerank_results(question, retrieved, top_k=K)

    # Calculate total relevant chunks in knowledge base
    total_relevant = sum(1 for c in chunks
                        if is_relevant(c.page_content, gt["relevant_keywords"]))
    total_relevant = max(total_relevant, 1)

    # Retrieval metrics
    p_at_k = precision_at_k(reranked, gt["relevant_keywords"], K)
    r_at_k = recall_at_k(reranked, gt["relevant_keywords"], K, total_relevant)
    mrr = mrr_score(reranked, gt["relevant_keywords"])

    # Generation metrics
    baseline_correct = correctness_score(baseline_ans, gt["correct_answer_keywords"])
    rag_correct = correctness_score(rag_ans, gt["correct_answer_keywords"])
    rag_grounding = grounding_score(rag_ans, reranked)

    evaluation_results.append({
        "question": question[:50] + "...",
        "precision_at_k": round(p_at_k, 2),
        "recall_at_k": round(r_at_k, 2),
        "mrr": round(mrr, 2),
        "baseline_correctness": round(baseline_correct, 2),
        "rag_correctness": round(rag_correct, 2),
        "rag_grounding": round(rag_grounding, 2)
    })

    print(f"\nQuestion: {question}")
    print(f"  Precision@{K}:          {p_at_k:.2f}")
    print(f"  Recall@{K}:             {r_at_k:.2f}")
    print(f"  MRR:                 {mrr:.2f}")
    print(f"  Baseline Correct:    {baseline_correct:.2f}")
    print(f"  RAG Correct:         {rag_correct:.2f}")
    print(f"  RAG Grounding:       {rag_grounding:.2f}")
    print("-" * 40)

# ============================================================
# STEP 4: Summary Table
# ============================================================

print("\n" + "=" * 70)
print("SUMMARY TABLE")
print("=" * 70)
print(f"{'Question':<45} {'P@K':>5} {'R@K':>5} {'MRR':>5} {'Base':>5} {'RAG':>5} {'Grnd':>5}")
print("-" * 70)

for r in evaluation_results:
    print(f"{r['question']:<45} "
          f"{r['precision_at_k']:>5.2f} "
          f"{r['recall_at_k']:>5.2f} "
          f"{r['mrr']:>5.2f} "
          f"{r['baseline_correctness']:>5.2f} "
          f"{r['rag_correctness']:>5.2f} "
          f"{r['rag_grounding']:>5.2f}")

# Averages
avg_precision = sum(r["precision_at_k"] for r in evaluation_results) / len(evaluation_results)
avg_recall = sum(r["recall_at_k"] for r in evaluation_results) / len(evaluation_results)
avg_mrr = sum(r["mrr"] for r in evaluation_results) / len(evaluation_results)
avg_baseline = sum(r["baseline_correctness"] for r in evaluation_results) / len(evaluation_results)
avg_rag = sum(r["rag_correctness"] for r in evaluation_results) / len(evaluation_results)
avg_grounding = sum(r["rag_grounding"] for r in evaluation_results) / len(evaluation_results)

print("-" * 70)
print(f"{'AVERAGE':<45} "
      f"{avg_precision:>5.2f} "
      f"{avg_recall:>5.2f} "
      f"{avg_mrr:>5.2f} "
      f"{avg_baseline:>5.2f} "
      f"{avg_rag:>5.2f} "
      f"{avg_grounding:>5.2f}")

print("\n" + "=" * 70)
print("FINAL SCORES")
print("=" * 70)
print(f"Average Precision@{K}:      {avg_precision:.2f}")
print(f"Average Recall@{K}:         {avg_recall:.2f}")
print(f"Average MRR:             {avg_mrr:.2f}")
print(f"Baseline Correctness:    {avg_baseline:.2f} ({avg_baseline*100:.0f}%)")
print(f"RAG Correctness:         {avg_rag:.2f} ({avg_rag*100:.0f}%)")
print(f"RAG Grounding Score:     {avg_grounding:.2f} ({avg_grounding*100:.0f}%)")
print(f"Improvement over Baseline: +{(avg_rag - avg_baseline)*100:.0f}%")

# Save results
json.dump(evaluation_results, open("evaluation_results.json", "w"), indent=2)
print("\nEvaluation results saved to evaluation_results.json")
```
